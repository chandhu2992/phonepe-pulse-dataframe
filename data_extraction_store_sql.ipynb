# Aggregated_Insurance

```python
#aggregated Insurance
path1 = "J:\\project\\phonepe\pulse\\data\\aggregated\\insurance\\country\\india\\state\\"
aggre_ins_list = os.listdir(path1)
columns1={"States":[], "Years":[], "Quarter":[], "Transaction_Type":[], "Transaction_Count":[], "Transactions_amount":[]}

for state in aggre_ins_list:
    current_states = path1 + state + "/"
    aggre_year_list=os.listdir(current_states)

    for year in aggre_year_list:
        current_year=current_states + year + "/"
        aggre_file_list=os.listdir(current_year)

        for file in aggre_file_list:
            current_file=current_year + file
            data=open(current_file,"r")
            A=json.load(data)
            for i in A["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                columns1["Transaction_Type"].append(name)
                columns1["Transaction_Count"].append(count)
                columns1["Transactions_amount"].append(amount)
                columns1["States"].append(state)
                columns1["Years"].append(year)
                columns1["Quarter"].append(int(file.strip(".json")))
```
